```python
import csv
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

## Regular Picking

```python
# getting relevant data into lists
with open('results/regular_picking.csv', 'r') as f:
    data = list(csv.reader(f, delimiter=","))
#social_move_vals = list(map(float, data[9][1:]))
data = list(map(float, data[11][1:]))
```

```python
picking_avgs = []
total_time_avgs = []
daily = []
for i in range(int((len(data))/20)):
    picking_avgs.append(sum(data[1 + 20*i:1 + 20*(i+1):2])/10)
    total_time_avgs.append(sum(data[20*i:20*(i+1):2])/10)
    daily.append(i + 1)
```

```python
plt.plot(daily, picking_avgs)
plt.xlabel("Average Mangroves Removed Daily")
plt.ylabel("Average Work Done")
```

```python
plt.plot(daily, total_time_avgs)
plt.xlabel("Average Mangroves Removed Daily")
plt.ylabel("Average Time To Remove All Mangroves")
```

## Big Effort

```python
# getting relevant data into lists
with open('results/big_effort.csv', 'r') as f:
    data = list(csv.reader(f, delimiter=","))
#social_move_vals = list(map(float, data[9][1:]))
data = list(map(float, data[13][1:]))
data_7 = data[:200]
data_30 = data[200:400]
data_90 = data[400:600]
data_180 = data[600:]
print(len(data_90))
periods = ["7", "30", "90", "180"]
```

```python
picking_avgs = [[],[],[],[]]
total_time_avgs = [[],[],[],[]]
daily = list(range(1,11))
for j in range(len(periods)):
    for i in range(int((len(data_7))/20)):
        picking_avgs[j].append(sum(data[200*j + 1 + 20*i:200*j + 1 + 20*(i+1):2])/10)
        total_time_avgs[j].append(sum(data[200*j + 20*i:200*j + 20*(i+1):2])/10)

print(picking_avgs)
print(total_time_avgs)
print(daily)
```

```python
for i in range(len(periods)):
    plt.plot(daily, picking_avgs[i])
    plt.xlabel("Average Mangroves Removed Daily")
    plt.ylabel("Average Work Done with Period = " + periods[i])
    plt.show()
```

```python

for i in range(len(periods)):
    plt.plot(daily, total_time_avgs[i])
    plt.xlabel("Average Mangroves Removed Daily")
    plt.ylabel("Average Time To Remove All with Period = " + periods[i])
    plt.show()
```
